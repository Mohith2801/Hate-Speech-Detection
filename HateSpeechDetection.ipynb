```python
dataset=pd.read_csv(r'C:\Users\narra\OneDrive\Desktop\projects\Data Science\Hate speech detection\twitter_data.csv')
dataset
```

```python
dataset['labels']=dataset['class'].map({0:"Hate speech detected", 1:"Offensive Language Detected", 3:"No hate and offensive speech"})
dataset
```

```python
ds=dataset[['tweet','labels']]
ds
```

```python
import re
import nltk
import string
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopword=set(stopwords.words("english"))
```

```python
stemmer=nltk.SnowballStemmer("english")
```

```python
def clean_data(text):
  text=str(text).lower()
  text=re.sub('https?://\S+|www\.\S+','',text)
  text=re.sub('\[.*?\]','',text)
  text=re.sub('<.*?>+','',text)
  text=re.sub('[%s]' %re.escape(string.punctuation),'',text)
  text=re.sub('\n','',text)
  text=re.sub('\w*\d\w*','',text)
  text=[word for word in text.split(' ') if word not in stopword]
  text=" ".join(text)
  text=[stemmer.stem(word) for word in text.split(' ')]
  text=" ".join(text)
  return text
```

```python
ds['tweet']=ds['tweet'].apply(clean_data)
```

```python
ds
```

```python
ds.isnull().sum()
```

```python
ds=ds.dropna()
```

```python
x=np.array(ds['tweet'])
y=np.array(ds['labels'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
cv=CountVectorizer()
x=cv.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf= DecisionTreeClassifier()
clf.fit(x_train, y_train)
```

```python
test_data='you are poor'
ds=cv.transform([test_data]).toarray()
print(clf.predict(ds))
```

```python
import pickle

# Save the CountVectorizer
with open("cv.pkl", "wb") as f:
    pickle.dump(cv, f)

# Save the trained classifier
with open("clf.pkl", "wb") as f:
    pickle.dump(clf, f)
```
